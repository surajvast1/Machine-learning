```python
# %%
import numpy as np 
import pandas as pd

df = pd.read_csv('petrolconsumption.csv')
df

# %%
print(df.isnull().sum())

# %% [markdown]
# Handling missing values ->no missing values
# Handling label encoding data -> no categorical values

# %%
numerical_cols = df.columns[[0,1,2]]  # Using all four columns as features
target_cols = df.columns[4]

# %%
# performing feature scaling on numerical data

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X = scaler.fit_transform(df[numerical_cols])
y = df[target_cols]

# %%
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

# %%
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,y_train)

# %% [markdown]
# Making predictions

# %%
y_pred = model.predict(X_test)

# %%
import matplotlib.pyplot as plt 

# Plotting the predictions vs actual values for one of the features
plt.scatter(X_test[:, 0], y_test, color='blue', label='Actual data')
plt.scatter(X_test[:, 0], y_pred, color='red', label='Predicted data')

plt.xlabel('Feature 1')
plt.ylabel('Petrol consumption')
plt.legend()
plt.show()

# %%
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(f'R2 Score: {r2}')
```
