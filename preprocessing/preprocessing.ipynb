```python
import pandas as pd
from sklearn.impute import SimpleImputer

df = pd.read_csv('car-data.csv')
df
```

```python
print(df.isnull().sum())
```

Seperating categorical and numerical values and hangling missing values 

```python
# Handling categroical missing values 
categorycal_cols = df.columns[[0,1,3]]
imputer = SimpleImputer(strategy='most_frequent')

df[categorycal_cols] = imputer.fit_transform(df[categorycal_cols])
```

```python
print(df[categorycal_cols] )
```

```python
# Handling numerical missing values
numerical_cols = df.columns[[2]]
imputer = SimpleImputer(strategy='mean')
df[numerical_cols] = imputer.fit_transform(df[numerical_cols])
```

```python
print(df.isnull().sum())
```

Now for the target column since it does not convert it is 1D array we need to reshape it and convert it to 2D array and pass it to the imputer 


```python
target_col = df.columns[[-1]]

y_label = df[target_col].values.reshape(-1,1)
imputer = SimpleImputer(strategy='mean')
y_imputed = imputer.fit_transform(y_label)
y_imputed = y_imputed.ravel()
target_col = y_imputed
```

```python
print(df.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder

for col in categorycal_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

df
```
