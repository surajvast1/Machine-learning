```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris['feature_names']
```

```python
iris['target_names']
```

```python
import pandas as pd
df = pd.DataFrame(data=iris["data"],columns=iris['feature_names'])
df['target'] = iris['target']
df
```

```python
# Correlation

correlation_cal = df.corr()['target'].abs().sort_values(ascending=True)
print(correlation_cal)
```

```python
# iris virginica or not


# Chososing two column one independent 
# most corellated one and other the target column 
# where we are transforming to a new frame where if virginica
# 1 else 0 
import numpy as np
X = iris["data"][:,3:]
# X
y=(iris["target"]==2).astype(np.int64)
y 
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()

clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
y_probability = clf.predict_proba(X_test)
```

```python
# Displaying the predicted classes
print("Predicted classes:", y_pred)

# Displaying the predicted probabilities
print("Predicted probabilities:")
print(y_probability)
```

```python
x_multi = iris["data"][:,[2,3]]
y_multi = iris["target"]
```

```python
X_train_multi,X_test_multi,y_train_multi,y_test_multi = train_test_split(x_multi,y_multi,test_size=0.25)
```

```python
softmax = LogisticRegression(multi_class="multinomial",solver="lbfgs")

softmax.fit(X_train_multi,y_train_multi)
```

```python
y_pred_multi = softmax.predict(X_test_multi)
y_prob_multi_= softmax.predict_proba(X_test_multi)
```

```python
print("Predicted classes (multi-class):", y_pred_multi)

# Displaying the predicted probabilities for multi-class logistic regression
print("Predicted probabilities (multi-class):")
print(y_prob_multi_)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test_multi,y_pred_multi))
```
